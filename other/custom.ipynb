```python
import sys, os
sys.path.append(os.path.abspath(".."))

import torch
from utils.graph_creation import create_hetero_graph
from utils.training_utils import train_val_test_split, train
from models.custom.custom import Custom
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
hetero_data = create_hetero_graph(holds_as_nodes=True)
```

```python
edge_type = ("user", "rates", "problem")
# 2) Split into message / train / val / test
message_data, train_data, val_data, test_data = train_val_test_split(
    hetero_data,
    edge_type=edge_type,
    message_p=0.7,
    train_p=0.1,
    val_p=0.1,
    by_user=True,
)
```

```python
# 3) Instantiate the model
model = Custom(
    hetero_data=message_data,   # same metadata and dims
    hidden_channels=64,
    num_layers=2,
    dropout=0.1,
)
```

```python
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3, weight_decay=1e-5)
```

```python
# 4) Train using your existing train() function
train(
    model=model,
    message_data=message_data,
    train_data=train_data,
    val_data=val_data,
    edge_type=edge_type,
    optimizer=optimizer,
    hetero=True,       # IMPORTANT: heterogeneous mode
    features=True,     # IMPORTANT: we do use features
    device=device,
    num_epochs=20,
    batch_size=1024,
)
```
