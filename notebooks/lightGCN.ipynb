```python
from utils.graph_creation import create_hetero_graph
from utils.training_utils import train_val_test_split_homogeneous, train
from torch_geometric.nn import LightGCN
import torch
```

```python
data = create_hetero_graph(holds_as_nodes=False)
message_data, train_data, val_data, test_data = train_val_test_split_homogeneous(
    data, edge_type=("user", "rates", "problem")
)
```

```python
model = LightGCN(
    num_nodes=message_data.x.shape[0],
    embedding_dim=64,
    num_layers=3,
)
```

```python
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

train(model, message_data, train_data, val_data, edge_type=None, optimizer=optimizer, 
      hetero = False, features=False, num_epochs=20)
```
